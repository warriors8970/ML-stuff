# Random Example

```python
np.random.seed(0)
sales_amt = np.random.randint(20,size = (5,3))
sales_amt
```

```python
weekly_sales = pd.DataFrame(sales_amt,index = ["mon","tue","wed","thurs","fri",],
                                       columns=["A1","P1","C1"])
```

```python
weekly_sales
```

```python
#creating price array
price = np.array([10,8,12])
price
```

```python
#price df
butter = price.reshape(1,3)
price_df = pd.DataFrame(butter,index=["Price"],
                           columns = ["A1","P1","C1"])
price_df
```

```python
total_sales = price.dot(sales_amt.T)
```

```python
total_sales
```

```python
butter.shape, weekly_sales.shape
```

```python
daily_sales = price_df.dot(weekly_sales.T)
type(daily_sales)
```

```python
daily_sales = daily_sales.T
```

```python
weekly_sales["total"] = daily_sales
weekly_sales
```
