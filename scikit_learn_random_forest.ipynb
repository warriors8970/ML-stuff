```python
import pandas as pd
heart_disease =  pd.read_csv("heart-disease.csv")
```

```python
heart_disease #import dataset - all numerical
```

```python
X = heart_disease.drop("target",axis = 1) #assign feature matrix
y = heart_disease["target"] #assign label matrix
```

```python
from sklearn.ensemble import RandomForestClassifier #a ml model
clf = RandomForestClassifier()

clf.get_params() #its parameters
```

```python
from sklearn.model_selection import train_test_split #splitting the data

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
clf.fit(X_train,y_train) #initializing the model
```

```python
 X_train
```

```python
X_test
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
#evaluating the model
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
import numpy as np
```

```python
#to improve this model, change n_estimators
np.random.seed(42)
for i in range(10,100,10):
    print(f"{i} estimators:")
    model = RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
    print(f"Model Accuracy : {model.score(X_test,y_test)*100 :.2f} %")
    print("")
```

```python
#save and import the model
import pickle
pickle.dump(clf,open("random_forest_model_1.pkl","wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pkl","rb"))
loaded_model.score(X_test,y_test)
```
